```python
import pandas as pd
df = pd.read_csv("./table-1.csv")
df = df.iloc[4:, :]
df.head()
```

```python
european = [
    "Austria", "Belgium", "Bulgaria", "Croatia", "Republic of Cyprus",
    "Czech Republic", "Denmark", "Estonia", "Finland", "France", "Germany",
    "Greece", "Hungary", "Ireland", "Italy", "Latvia", "Lithuania",
    "Luxembourg", "Malta", "Netherlands", "Poland", "Portugal", "Romania",
    "Slovakia", "Slovenia", "Spain", "Sweden"
]
```

```python
data = pd.DataFrame({"State":df.Date, "Jan": df.iloc[:, 2], "Feb": df.iloc[:, 3],
                   "Mar": df.iloc[:, 4], "Apr": df.iloc[:, 5], "May": df.iloc[:, 6],
                   "Jun": df.iloc[:, 7], "Jul": df.iloc[:, 8], "Aug": df.iloc[:, 9],
                   "Sep": df.iloc[:, 10], "Oct": df.iloc[:, 11], "Nov": df.iloc[:, 12]})
data
```

```python
data = data[data['State'].isin(european)]
data.to_csv("data.csv")
```
